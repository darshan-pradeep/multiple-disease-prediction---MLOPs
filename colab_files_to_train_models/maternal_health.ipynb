```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
maternal_health_risk = fetch_ucirepo(id=863)

# data (as pandas dataframes)
X = maternal_health_risk.data.features
y = maternal_health_risk.data.targets

# metadata
print(maternal_health_risk.metadata)

# variable information
print(maternal_health_risk.variables)
```

# Task
Train a RandomForestClassifier model using `X` and `y` data. Split the data into training and testing sets, then train the model and evaluate its accuracy.

## Import Libraries

### Subtask:
Import necessary libraries for model training and evaluation.


**Reasoning**:
Importing the specified libraries for model training and evaluation, as instructed, will make them available for use in subsequent steps.



```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

print("Libraries imported successfully.")
```

## Split Data

### Subtask:
Split the `X` and `y` data into training and testing sets using `train_test_split`. Also, ensure `y` is a 1D array for model training.


**Reasoning**:
The subtask requires splitting the data into training and testing sets and ensuring the target variable `y` is a 1D array. The code will convert `y` to a 1D array and then use `train_test_split` to divide the data.



```python
y_raveled = y.values.ravel()
X_train, X_test, y_train, y_test = train_test_split(X, y_raveled, test_size=0.2, random_state=42)

print("Data split into training and testing sets successfully.")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

## Train Model

### Subtask:
Initialize a `RandomForestClassifier` and train it using the training data.


**Reasoning**:
The subtask requires initializing a `RandomForestClassifier` and training it with the provided training data. This code will create a model instance and then fit it to `X_train` and `y_train`.



```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

print("RandomForestClassifier model trained successfully.")
```

## Make Predictions

### Subtask:
Use the trained model to make predictions on the test set.


**Reasoning**:
The subtask is to make predictions using the trained model on the test set. I will use the `predict` method of the trained model (`model`) on the `X_test` data and store the results in `y_pred`.



```python
y_pred = model.predict(X_test)

print("Predictions made successfully.")
print(f"Shape of y_pred: {y_pred.shape}")
```

## Evaluate Model

### Subtask:
Evaluate the accuracy of the model using the `accuracy_score` metric.

**Reasoning**:
The subtask requires evaluating the accuracy of the model. I will use the `accuracy_score` function to compare the predicted values (`y_pred`) with the actual test values (`y_test`).



```python
accuracy = accuracy_score(y_test, y_pred)

print(f"Model Accuracy: {accuracy:.4f}")
```

Generate and print a classification report from `sklearn.metrics` to evaluate model performance using `y_test` and `y_pred`.

### Subtask:
Import the `classification_report` function from `sklearn.metrics`.

```python
from sklearn.metrics import classification_report

print("classification_report imported successfully.")
```

The subtask is to print the classification report. I will use the imported `classification_report` function with `y_test` and `y_pred` to generate and print the report.
